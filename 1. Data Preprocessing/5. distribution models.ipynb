```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
```

```python
# normal distribution
np.random.seed(1234)
```

```python
# This generates 10,000 samples from a log-normal distribution with a specified mean
# and sigma (standard deviation) for the underlying normal distribution.
# mean=1 and sigma=0.4 are the parameters for the normal distribution before exponentiation.
samples = np.random.lognormal(mean=1, sigma=0.4, size=10000)
```

```python
sample,loc,scale = scipy.stats.lognorm.fit(samples, floc=0)
```

```python
num_bins=50
```

```python
counts,edges,patches = plt.hist(samples, bins=num_bins, color='r')
```

```python
# binomial distribution
import seaborn
from scipy.stats import binom
```

```python
data = binom.rvs(n=17, p=0.7, loc=0, size=1010)
```

```python
ax = seaborn.distplot(data, kde=True, color='pink', hist_kws={'linewidth':22,'alpha':0.77})
```
